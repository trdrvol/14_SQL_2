```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sqlite3
```

```python
import pandasql as sqldf
pysqldf = lambda q : sqldf(q, globals())
```

```python
Adult = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", names= ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','class'])
def TrimColumns(df):
    trim_all= lambda x : x.strip() if type(x) is str else x
    return df.applymap(trim_all)
Adult = TrimColumns(Adult)

Adult.columns=Adult.columns.str.replace("-","_")
Adult.head()
    
```

```python
# We start engine by following command
```

```python
import sqlalchemy
from sqlalchemy import create_engine
engine=create_engine("sqlite:///SQLADB", echo = False)
```

```python
Adult.to_sql("SQLADB",engine, if_exists="replace")
```

```python
connection = sqlite3.connect("SQLADB")
cursor= connection.cursor()
```

```python
# Write 2 basics update queries

cursor.execute("UPDATE SQLADB SET  fnlwgt = '11111' WHERE fnlwgt = '77516';")
pd.read_sql_query("SELECT * FROM SQLADB", connection).head(3)
```

```python
cursor.execute("UPDATE SQLADB SET marital_status = 'Divorced' WHERE age='50' and education_num='13';")
pd.read_sql_query("SELECT * FROM SQLADB",connection).head(4)
```
